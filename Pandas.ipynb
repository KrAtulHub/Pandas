```python
import pandas as pd 
import numpy as np 

data ={
    "Name": ['Alice','Bob','Charlie','David','Emma'], 
    "Math": [88,75,92,60,85], 
    "Science": [90,80,78,70,95]
}

df =pd.DataFrame(data)
```

```python
df.head()
```

```python
df['Total_Marks'] =df[['Math','Science']].apply(lambda row: row.sum(),axis =1)
```

```python
df
```

```python
col_mean =df[['Math','Science']].apply(lambda col:col.mean())
```

```python
col_mean
```

```python
def to_grade(s): 
    if s>=90: 
        return "A"
    elif s>=80: 
        return "B"
    else:
        return "C"
df['Math_Grade'] =df['Math'].apply(to_grade)
```

```python
df
```

```python
df['Science_Grade'] =df['Science'].apply(to_grade)
```
